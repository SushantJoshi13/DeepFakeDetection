```python
%matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
from skimage import io
from skimage.feature import graycomatrix, graycoprops
```

```python
video_folder = "Reduced_Dataset"
```

```python
imgPath = []
for label in ["Real","Fake"]:
        video_label_folder = os.path.join(video_folder, label)
        for video_filename in os.listdir(video_label_folder):
            imgList = os.path.join(video_label_folder, video_filename)
            for path in os.listdir(imgList):
                pathImg = os.path.join(imgList,path)
                imgPath.append(pathImg)
imgPath.sort()
print(imgPath[1])
```

```python
def is_image_by_extension(filename):
  extensions = (".jpg", ".jpeg", ".png", ".bmp", ".gif", ".tiff")  # Add more extensions as needed
  return filename.lower().endswith(extensions)
```

```python
av = []
aw = []
ax = []
ay = []
az = []
for patch in imgPath:
    # print(patch)
    if is_image_by_extension(patch):
        image = io.imread(patch)
        glcm = graycomatrix(image, distances=[1], angles=[0], levels=256)
        av.append(graycoprops(glcm, 'dissimilarity')[0, 0])
        aw.append(graycoprops(glcm, 'correlation')[0, 0])
        ax.append(graycoprops(glcm, 'contrast')[0, 0])
        ay.append(graycoprops(glcm, 'homogeneity')[0, 0])
        az.append(graycoprops(glcm, 'energy')[0, 0])
```

```python

imgName = []
for label in ["Real","Fake"]:
        video_label_folder = os.path.join(video_folder, label)
        for video_filename in os.listdir(video_label_folder):
            imgList = os.path.join(video_label_folder, video_filename)
            for path in os.listdir(imgList):
                imgName.append(path)
imgName.sort()
print(imgName[100])
```

```python
import os

imgNames = []
labels = []

for label in ["Real", "Fake"]:
    video_label_folder = os.path.join(video_folder, label)
    for video_filename in os.listdir(video_label_folder):
        imgList = os.path.join(video_label_folder, video_filename)
        for path in os.listdir(imgList):
            imgNames.append(path)
            labels.append(0 if label == "Real" else 1)  # Assign 0 for "Real" and 1 for "Fake"

# Sort the image names and labels together based on image names
sorted_data = sorted(zip(imgNames, labels), key=lambda x: x[0])

# Extract the first element after sorting
print(sorted_data[1])
```

```python
submission = pd.DataFrame(dict(label = labels,image_name=imgNames, av=av,aw=aw,ax=ax, ay=ay,az=az, bv=bv,bw=bw,bx=bx, by=by,bz=bz, cv=cv,cw=cw,cx=cx, cy=cy,cz=cz, dv=dv,dw=dw,dx=dx, dy=dy,dz=dz, ev=ev,ew=ew,ex=ex, ey=ey,ez=ez))
submission = submission.sort_values('image_name')
submission.to_csv('feature1.csv', index=False)
submission.head()
```
